1. Find the largest, smallest and second largest number from 10 user inputs.

```python
number=[]
for i in range(10):
    num =input ("Enter number : ")
    number.append(int(num))
print("number : ",number)

largest=number[0]
smallest=number[0]

for n in number:
    if n >largest:
        largest =n
    if n<smallest:
        smallest=n
second_largest=smallest
for n in number:
    if n> second_largest and n!=largest:
        second_largest= n
            
print("largest : ",largest)
print("smallest : ",smallest)
print("second largest : ",second_largest)

```

2. Count vowels, consonants, digits, and special characters in a string.

```python
s = input("enter a string: ")
vowels =0
consonants =0
digits =0
special = 0
for i in s:
    if i.isalpha():
        if i.lower() in "aeiou":
            vowels += 1
        else:
            consonants += 1
    elif i.isdigit():
        digits += 1
    else:
        special += 1
print(s)
print("vowels : ", vowels)
print("consonants : ", consonants)
print("digits : ", digits)
print("special characters : ", special)
```

3. Check whether a number is an Armstrong number.

```python
num =int(input("enter numer : "))
num_copy=num
power=len(str(num))

sum =0
while num_copy >0:
    digit=num_copy % 10
    sum += digit**power
    num_copy=num_copy//10
    
print("num",num)
print("sum",sum)
if num == sum :
    print("number is an Armstrong number")
else:
    print("number is not an Armstrong number")
```

4. Find duplicate elements and their frequency in a tuple.

```python
t = (1, 2, 3, 9,9,6, 4, 5, 1, 2, 6, 1)
frequencies = {}
for element in t:
    if element in frequencies:
        frequencies[element]+=1
    else:
        frequencies[element]=1
print(t)
for element,count in frequencies.items():
    if count>1:
        print (f"element {element} is repeat in {count} times" )
```

7. Display students scoring above the average marks from a dictionary.

```python
student = {
    "krishan": 50,
    "kalpesh": 80,
    "lakshit": 70,
    "ankush": 60,
    "ajit": 90,
    "ashish": 60
}
sum = 0
count = 0
for name, marks in student.items():
    sum += marks
    count += 1
average = sum / count
print(student)
print(f"The average mark is: {average}")
print("Students scoring above average:")
for name, marks in student.items():
    if marks > average:
        print(f"Student: {name} with Marks: {marks}")
```

8. Create a menu-driven product management system using a dictionary.

```python
products = {}

while True:
    print("===== Product Management System =====")
    print("1. add product")
    print("2. view products")
    print("3. update product")
    print("4. delete product")
    print("5. search product")
    print("6. exit")

    choice = int(input("enter your choice: "))

    if choice == 1:
        name = input("enter product name: ")
        price = float(input("enter product price: "))
        products[name] = price
        print("product added successfully!")

    elif choice == 2:
        if products:
            print("\nproduct List:")
            for name, price in products.items():
                print(f"{name} : ₹{price}")
        else:
            print("no products available.")

    elif choice == 3:
        name = input("enter product name to update: ")
        if name in products:
            new_price = float(input("enter new price: "))
            products[name] = new_price
            print("product updated successfully!")
        else:
            print("product not found.")

    elif choice == 4:
        name = input("enter product name to delete: ")
        if name in products:
            del products[name]
            print("product deleted successfully!")
        else:
            print("product not found.")

    elif choice == 5:
        name = input("enter product name to search: ")
        if name in products:
            print(f"{name} : ₹{products[name]}")
        else:
            print("product not found.")

    elif choice == 6:
        print("Exiting Product Management System...")
        break

    else:
        print("invalid choice.")
```

9. Check whether a triangle is Equilateral, Isosceles, or Scalene.

```python
a = int(input("Enter first side: "))
b = int(input("Enter second side: "))
c = int(input("Enter third side: "))

if a == b and b == c:
    print("Equilateral Triangle")

elif a == b or b == c or a == c:
    print("Isosceles Triangle")

else:
    print("Scalene Triangle")
```

10. Create a Student Management System using loops, tuples, dictionaries, and conditions.

```python
students = {}
while True:
    print("\n--- Student Management System ---")
    print("1. Add Student")
    print("2. View Students")
    print("3. Search Student")
    print("4. Delete Student")
    print("5. Exit")

    ch = int(input("Enter your choice: "))

    if ch == 1:
        roll = int(input("Enter Roll No: "))
        name = input("Enter Name: ")
        marks = int(input("Enter Marks: "))

        # tuple to store student data
        students[roll] = (name, marks)

        print("Student added successfully")

    elif ch == 2:
        if len(students) == 0:
            print("No student records found")
        else:
            print("\nStudent Records:")
            for roll, data in students.items():
                print("Roll No:", roll)
                print("Name:", data[0])
                print("Marks:", data[1])

                if data[1] >= 40:
                    print("Result: Pass")
                else:
                    print("Result: Fail")
                print()

    elif ch == 3:
        r = int(input("Enter Roll No to search: "))

        if r in students:
            print("Name:", students[r][0])
            print("Marks:", students[r][1])
        else:
            print("Student not found")

    elif ch == 4:
        r = int(input("Enter Roll No to delete: "))

        if r in students:
            del students[r]
            print("Record deleted")
        else:
            print("Student not found")

    elif ch == 5:
        print("Program Ended")
        break

    else:
        print("Invalid Choice")
```
